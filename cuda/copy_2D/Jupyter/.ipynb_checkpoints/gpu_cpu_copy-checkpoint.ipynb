# Performance Analysis of GPU Copy Functions

## Introduction

In this notebook, we analyze the performance of GPU copy. We will plot the time elapsed for these operations against the number of elements processed.

```python
import matplotlib.pyplot as plt
import mygputest.image_copy as image_copy
```

## Results

The following plot shows the relationship between the time elapsed and the number of elements processed.

```python
# Initialize variables
num_treads_per_block = 512
time_elapsed_gpu = []
time_elapsed_cpu = []
num_elements_list = []
num_plot_points=1000
step_plot_points=10
min_num=10

# Collect data for num_treads_per_block = 512
image_copy.enableDevice();
image_copy.inititalize(min_num+(num_plot_points*step_plot_points*num_treads_per_block))
image_copy.MoveImageToDevice(Pixel* data);
image_copy.image_copy();
image_copy.GetImageFromDevice(Pixel* data);
image_copy.clear_all();
```

## Results

The following plot shows the relationship between the time elapsed and the number of elements processed.

```python
# Plotting the data
plt.figure(figsize=(12, 8))

# Plot time_elapsed for 512 threads per block
plt.plot(num_elements_list, time_elapsed_gpu, label='512 threads/block', marker='o')
plt.plot(num_elements_list, time_elapsed_cpu, label='std::copy', marker='o')

# Adding titles and labels
plt.title('Time Elapsed vs Number of Elements')
plt.xlabel('Number of Elements')
plt.ylabel('Time Elapsed (ms)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

*Figure 1: Time elapsed for GPU copy functions with and without shared memory for different numbers of threads per block.*
